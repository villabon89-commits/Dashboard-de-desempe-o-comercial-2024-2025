### Narrativa del Dashboard (Modelo SCQA)

#### 🖥️ Vista General (Overview)

**S (Situación):** Andes Retail Group ha experimentado un período de crecimiento, sin embargo, los datos más recientes indican una desaceleración en el rendimiento general.

**C (Complicación):** Hemos observado una caída interanual significativa del -6.80% en los ingresos totales entre 2024 y 2025.

**Q (Pregunta):** ¿Cuáles son los principales factores que contribuyen a esta disminución de ingresos a nivel macro y cómo se distribuye esta caída?

**A (Respuesta):** La caída se debe principalmente a una desaceleración notable en nuestro mercado clave, Perú, y una disminución en la contribución del sector Premium y la categoría de Electrónica.

---

#### 🔎 Vista Detalle

**S (Situación):** El análisis granular de los ingresos revela puntos específicos donde la disminución ha sido más pronunciada, afectando el rendimiento general.

**C (Complicación):** Perú, nuestro mercado principal, mostró una desaceleración bajando de 1.14M a 1.02M. El sector Premium perdió 0.20M y la categoría de Electrónica se contrajo en 0.09M. Adicionalmente, las ventas sufren un impacto crítico y recurrente durante los meses de invierno.

**Q (Pregunta):** ¿Cómo podemos mitigar las pérdidas en mercados, segmentos y categorías específicas, y desarrollar estrategias para contrarrestar el impacto estacional en los ingresos?

**A (Respuesta):** Se requiere un plan de acción específico para Perú, incluyendo estrategias de marketing y ventas localizadas. Para el sector Premium y Electrónica, se debe revisar la oferta de productos, los precios y las campañas de promoción para revitalizar la demanda. Para la estacionalidad de invierno, se pueden explorar promociones específicas, diversificar la oferta de productos o lanzar campañas anticipadas para suavizar la caída de ventas.
